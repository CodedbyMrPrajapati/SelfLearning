G_0 = -1 + gamma(-1) + gamma^2(-1) + gamma^3(10)

## Task 2: Bellman Expectation Equation for \( v_\pi(S_2) \)

### Possible transitions from S2

| Action | Next State | Reward |
|------|-----------|--------|
| Left | S1 | -1 |
| Right | STerm | +10 |

### Bellman equation

\[
v_\pi(S_2)
= \frac{1}{2}(-1 + \gamma v_\pi(S_1))
+ \frac{1}{2}(10 + \gamma v_\pi(S_{Term}))
\]

Since STerm is terminal:

\[
v_\pi(S_{Term}) = 0
\]

### Final form

\[
v_\pi(S_2)
= \frac{1}{2}(-1 + 0.9\,v_\pi(S_1)) + 5
\]

### Python representation

```python
def v_pi_S2(v_S1, gamma=0.9):
    return 0.5 * (-1 + gamma * v_S1) + 5
```
